```python
import pandas as pd
import numpy as np

def preprocessing(test):
    
    #DESCRIPTION: the function processes a test set that is applied on a model learned on train_set
    #REMARQUE: the function might need to be adapted in function of the features that are used (some missing: third parties,...)
    
    #Replace claim_vehicle_date_inuse of rows considered as outliers (e.g. year 2700) by the median in the training set
    AUX=pd.DataFrame()
    AUX['YearMonth']=test['claim_date_occured']/100
    test.loc[test['claim_vehicle_date_inuse']>AUX['YearMonth'],'claim_vehicle_date_inuse']=201006
    
    
    #Date formats    
    column1 = ['claim_date_registered',
               'claim_date_occured'
              ]
    for i in column1:
        test[i] = pd.to_datetime(test[i], format='%Y%m%d')
        
    column2 = ['policy_date_start', 
               'policy_date_next_expiry', 
               'policy_date_last_renewed', 
               'claim_vehicle_date_inuse'
              ]
    for i in column2:
        test[i] = pd.to_datetime(test[i], format='%Y%m')
        
    column3 = ['policy_holder_year_birth', 
               'driver_year_birth', 
               'repair_year_birth', 
               'third_party_1_year_birth', 
               'third_party_2_year_birth', 
               'third_party_3_year_birth'
              ]
    for i in column3:
        test[i] = pd.to_datetime(test[i], format='%Y')    

    
    
    #New variables
    test['time_to_register']=(test['claim_date_registered'] - test['claim_date_occured'])
    test['time_to_register']=test['time_to_register']/np.timedelta64(1,'D')

    test['absence_time']=test['claim_time_occured'].isnull().values
    
    test['absence_alcohol_test']=test['claim_alcohol'].isnull().values
    
    test['absence_vehicle_id']=test['claim_vehicle_id'].isnull().values
    
    test['vehicle_age']=(test['claim_date_occured'] - test['claim_vehicle_date_inuse'])
    test['vehicle_age']=test['vehicle_age']/np.timedelta64(1,'Y')
    
    test['repair_store_age']=(test['claim_date_occured'] - test['policy_holder_year_birth'])
    test['repair_store_age']=test['repair_store_age']/np.timedelta64(1,'Y')
    
    test['absence_repair_id']=test['repair_id'].isnull().values
    
    test['absence_holder_experd_id']=test['policy_holder_expert_id'].isnull().values
    
    test['absence_driver_experd_id']=test['driver_expert_id'].isnull().values
    
    test['time_as_policy_holder']=(test['claim_date_occured'] - test['policy_date_start'])
    test['time_as_policy_holder']=test['time_as_policy_holder']/np.timedelta64(1,'Y')
    
    test['remaining_time_policy']=(test['policy_date_next_expiry'] - test['claim_date_occured'])
    test['remaining_time_policy']=test['remaining_time_policy']/np.timedelta64(1,'Y')
    
    test['time_policy_renewed']=(test['claim_date_occured'] - test['policy_date_last_renewed'])
    test['time_policy_renewed']=test['time_policy_renewed']/np.timedelta64(1,'Y')
    
    test['policy_holder_age']=(test['claim_date_occured'] - test['policy_holder_year_birth'])
    test['policy_holder_age']=test['policy_holder_age']/np.timedelta64(1,'Y')
    
    test['driver_age']=(test['claim_date_occured'] - test['driver_year_birth'])
    test['driver_age']=test['driver_age']/np.timedelta64(1,'Y')
    
    test['third_party_1_age']=(test['claim_date_occured'] - test['third_party_1_year_birth'])
    test['third_party_1_age']=test['third_party_1_age']/np.timedelta64(1,'Y')
    
    test['third_party_2_age']=(test['claim_date_occured'] - test['third_party_2_year_birth'])
    test['third_party_2_age']=test['third_party_2_age']/np.timedelta64(1,'Y')
    
    test['third_party_3_age']=(test['claim_date_occured'] - test['third_party_3_year_birth'])
    test['third_party_3_age']=test['third_party_3_age']/np.timedelta64(1,'Y')


    
    #Filling NaN values with the most frequent value in training
    test['policy_holder_postal_code']=test['policy_holder_postal_code'].fillna(4219)
    test['driver_postal_code']=test['driver_postal_code'].fillna(4020)
    test['repair_postal_code']=test['repair_postal_code'].fillna(4020)
    
    test['claim_postal_code']=test['claim_postal_code'].astype(int)
    test['policy_holder_postal_code']=test['policy_holder_postal_code'].astype(int)
    test['driver_postal_code']=test['driver_postal_code'].astype(int)
    test['repair_postal_code']=test['repair_postal_code'].astype(int)

    test['claim_language']=test['claim_language'].fillna(2.0)
    test['claim_language'] = test['claim_language'].astype(int)
    test['claim_language'] = test['claim_language'].astype(object)

    test['vehicle_age']=test['vehicle_age'].fillna(7)
    test['vehicle_age']=test['vehicle_age'].astype(int)
    
    test['third_party_1_age']=test['third_party_1_age'].fillna(48.08)
    test['third_party_1_age']=test['third_party_1_age'].astype(int)
    
    test['third_party_2_age']=test['third_party_2_age'].fillna(43.62)
    test['third_party_2_age']=test['third_party_2_age'].astype(int)
    
    test['third_party_3_age']=test['third_party_3_age'].fillna(57.32)
    test['third_party_3_age']=test['third_party_3_age'].astype(int)
    
    test['claim_vehicle_type']=test['claim_vehicle_type'].fillna('car')

    test['policy_holder_age']=test['policy_holder_age'].fillna(44.87)
    test['driver_age']=test['driver_age'].fillna(49.69)

    test['repair_country']=test['repair_country'].fillna('B')
    test['repair_store_age']=test['repair_store_age'].fillna(44.87)
    
    
    return test
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lehtlass/advanced-analytics/main/data/validation_set.csv", sep=';')
```

```python
df = preprocessing(df)
```
